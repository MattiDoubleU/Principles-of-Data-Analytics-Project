```python
# Numerical arrays and random numbers.
import numpy as np

# Plot.
import matplotlib.pyplot as plt
```

## Generate Random Integers
***

```python
rng = np.random.default_rng()
```

```python
# Generate one random float uniformly distributed over the range [0, 1)

rng.random()  
```

```python
# Generate an array of 10 numbers according to a unit Gaussian distribution.

rng.standard_normal(12)  
```

```python
# Generate an array of 5 integers uniformly over the range [0, 10).
rng.integers(low=0, high=10, size=100)  
```

```python
# Generate 10,000 integers over the range [0, 10).
x = rng.integers(low=0, high=10, size=10000, endpoint=True)  

#Show x.
x
```

```python
# Get the counts.
values, counts = np.unique(x, return_counts = True)

# Show.
values, counts
```

```python
# Plot the bar chart.
plt.bar(values, counts)
```

### Uniform Distribution
***


```python
# Generate 100 floats uniformly distributed over the range [0, 10).
x = rng.uniform(low=0, high=10, size=100)

# Show.
x
```

```python
# Get the counts.
values, counts = np.unique(x, return_counts = True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot the bar chart.
plt.bar(values, counts, edgecolor='black')
```

```python
# Round values in x.
x_rounded = np.round(x, decimals=0)

# Get the counts.
values, counts = np.unique(x_rounded, return_counts = True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot the bar chart.
plt.bar(values, counts, edgecolor='black')
```

```python
# Generate 10,000 floats uniformly distributed over the range [0, 10).
x = rng.uniform(low=0, high=10, size=10000)

# Show.
x
```

```python
# Get the counts.
values, counts = np.unique(x, return_counts = True)

# Show.
values, counts, values.size, counts.size
```

```python
# Plot histogram
plt.hist(x, edgecolor='black')
```

## Standard Normal
***

```python
# Generate an array of 10000 numbers.
x = rng.standard_normal(size=10000)

# Show
x
```

```python
# Plot histogram
plt.hist(x, edgecolor='red')
```
